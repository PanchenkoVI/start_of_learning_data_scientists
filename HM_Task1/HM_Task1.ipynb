```python
"""
Задание
+Собрать информацию о всех строящихся объектах на сайте "наш.дом.рф"
+Cохранить ее в pandas dataframe
+Cохранить pandas dataframe в excel
+Cохранить pandas dataframe в pickle
+Cохранить pandas dataframe в БД
-Задание *
Написать скрипт, который скачивает фотографии всех строящихся объектов с сайта "наш.дом.рф"
"""
import sqlite3
import requests 
from tqdm.auto import tqdm
import pandas as pd
import time
```

```python
offset_ = 0
n = 1077
limit_ = 10
all_data = []

for i in tqdm(range(n)):
    objects_data = []
    objects_list = []
    url = f'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/kn/object?offset={offset_}&sortField=devId.devShortCleanNm&sortType=asc&objStatus=0'
    res = ''
    while res == '':
        try:
            res = requests.get(url)
            objects_data = res.json()
            objects_data.get('data').get('list')
            objects_list = objects_data.get('data').get('list')
            all_data.extend(objects_list)
            for j in objects_list:
                tmp = [j.get('objId')]
                all_id.extend(tmp)
                all_id = list(set(all_id))
            offset_ = offset_ + limit_
            break
        except:
            print(f"ERROR - {i}")
            time.sleep(5)
            n = n + 1
            continue

print(len(all_data))
print(len(all_id))
df = pd.DataFrame(all_data)
df = df.sort_values(by=['objId'])
df.head(50)
```

```python
offset_ = 0
n = len(all_id)
limit_ = 10
all_data = []
rez = pd.DataFrame()

for i in tqdm(range(n)):
    objects_data = []
    objects_list = []
    tmp = []
    url = f'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/object/{all_id[i]}'
    res = ''
    while res == '':
        try:
            res = requests.get(url)
            objects_data = res.json()
            tmp = pd.json_normalize(objects_data.get('data'))
            rez = rez.append(tmp)
            offset_ = offset_ + limit_
            break
        except:
            print(f"ERROR - {i}")
            time.sleep(5)
            n + 1
            continue
len(rez)
```
