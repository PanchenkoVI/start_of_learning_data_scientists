```python
"""
Задание
+Собрать информацию о всех строящихся объектах на сайте "наш.дом.рф"
+Cохранить ее в pandas dataframe
+Cохранить pandas dataframe в excel
+Cохранить pandas dataframe в pickle
+Cохранить pandas dataframe в БД
-Задание *
Написать скрипт, который скачивает фотографии всех строящихся объектов с сайта "наш.дом.рф"
"""
import sqlite3
import requests 
from tqdm.auto import tqdm
import pandas as pd
import time
```

```python
offset_ = 0
n = 1078
limit_ = 20
all_data = []

for i in tqdm(range(n)):
    objects_data = []
    objects_list = []
    url = f'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/kn/object?offset={offset_}&sortField=devId.devShortCleanNm&sortType=asc&objStatus=0'
    res = ''
    while res == '':
        try:
            res = requests.get(url)
            break
        except:
            print(f"Connection refused by the server..{n}")
            time.sleep(5)
            print(f"Was a nice sleep, now let me continue...{n}")
            continue
    res = requests.get(url)
    objects_data = res.json()
    objects_data.get('data').get('list')
    objects_list = objects_data.get('data').get('list')
    all_data.extend(objects_list)
    offset_ = offset_ + 1

len(all_data)
```

```python
df = pd.DataFrame(all_data)
df.head()
```
