```python
class DomIdLoader:
    
    def __init__(self):
        self.offset_ = 0
        self.limit_ = 10
        self.all_data = []
        self.all_id = []
        self.cnt = 1000 #1077

    def get_some_data(self):
        
        for i in tqdm(range(self.cnt)):
            objects_data = []
            objects_list = []
            url = f'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/kn/object?offset={self.offset_}&sortField=devId.devShortCleanNm&sortType=asc&objStatus=0'
            res = ''
            while res == '':
                try:
                    res = requests.get(url)
                    objects_data = res.json()
                    objects_data.get('data').get('list')
                    objects_list = objects_data.get('data').get('list')
                    self.all_data.extend(objects_list)
                    for j in objects_list:
                        tmp = [j.get('objId')]
                        self.all_id.extend(tmp)
                        self.all_id = list(set(self.all_id))
                    self.offset_ = self.offset_ + self.limit_
                    break
                except:
                    print(f"ERROR - {i}")
                    time.sleep(5)
                    self.cnt = self.cnt + 1
                    continue
        df = pd.DataFrame(self.all_data)
        return df
        
    def get_a_lot_of_data(self):
        self.get_some_data()
        n = len(self.all_id)
        self.all_data = []
        rez = pd.DataFrame()

        for i in tqdm(range(n)):
            objects_data = []
            objects_list = []
            tmp = []
            url = f'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/object/{self.all_id[i]}'
            res = ''
            while res == '':
                try:
                    res = requests.get(url)
                    objects_data = res.json()
                    tmp = pd.json_normalize(objects_data.get('data'))
                    rez = rez.append(tmp)
                    break
                except:
                    print(f"ERROR - {i}")
                    time.sleep(5)
                    n + 1
                    continue
        return rez

if __name__ == '__main__':
    DIL = DomIdLoader()
    
df1 = DIL.get_some_data()       # Если необходимо сделать быстрый просмотр. (25 столбцов)
df2 = DIL.get_a_lot_of_data()   # Получить полную детальную информацию по объектам (более 90 столбцов)
```

```python
class Saver():
    def save_csv(self,data):
        data.to_csv('bd3.csv')
        
    def save_pickle(self,data):
        data.to_pickle('bd3.pkl')

    def save_xl(self, data):
        data.to_excel('bd.xlsx', encoding='utf8')

    def save_sql(self, data):
        df_sql = data.applymap(str)
        conn = sqlite3.connect('bd3.db')
        df_sql.to_sql('bd3', conn, if_exists='replace', index = False)
        conn.commit()
        conn.close()
        
        
if __name__ == '__main__':
    S = Saver()

S.save_csv(df2)
S.save_pickle(df2)
S.save_xl(df2)
S.save_sql(df2)
```

```python
class Visualizer():
    def __init__(self,data):
        self.data = data

    def make_boxplot_mean(self):
        data1 = self.data.groupby(['developer.regRegionDesc'], as_index=False)['objPriceAvg'].mean()
        data1 = data1.dropna(subset=['objPriceAvg'])
        data1 = data1.sort_values(by='objPriceAvg', ascending=True)
        data1 = data1.rename(columns={'objPriceAvg':'Средняя цена по рынку'})
        data1 = data1.sort_values(by='Средняя цена по рынку', ascending=False)
        rcParams['figure.figsize'] = 13,13
        data1[:3].plot(kind='barh')
        plt.xlabel('Цена')
        plt.ylabel('Номер региона')
        plt.show()

    def make_heatmap(self):
        
        data2 = self.data.groupby(['region'], as_index=False)['objPriceAvg'].mean()
        data2 = data2.dropna(subset=['objPriceAvg'])
        data2 = data2.rename(columns={'objPriceAvg':'Средняя цена по рынку'})
        data2 = data2.sort_values(by='Средняя цена по рынку', ascending=False)
        data2['Средняя цена по рынку'] = data2['Средняя цена по рынку'].apply(lambda row: math.ceil(row))
        
        data2.drop(columns=df.columns[1], axis=1, inplace=True)
#         data_set = np.random.rand(10,10)
        ax = sns.heatmap( data2 , linewidth = 0.5 , cmap = 'coolwarm' )
        plt.title( "Средняя цена по рынку" )
        plt.show()

if __name__ == '__main__':
    V = Visualizer(df)

V.make_boxplot_mean()
V.make_heatmap()

# Еще пларнирую править.. подробный разбор на графиках в во втором домашнем задании.
```
